```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("abhishek14398/salary-dataset-simple-linear-regression")

print("Path to dataset files:", path)
```

```python
import os
files = os.listdir(path)
print("Files in folder:", files)

csv_file = [f for f in files if f.endswith(".csv")][0]  # Get the CSV file
csv_path = os.path.join(path, csv_file)
```

```python
import pandas as pd
data = pd.read_csv(csv_path)
print(data.head())
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

x = data["YearsExperience"]
y = data["Salary"]

model = LinearRegression()
x = x.values.reshape(-1, 1)  # Reshape for sklearn which expects 2D array for features
y = y.values.reshape(-1, 1)

model.fit(x, y)
print("Coefficient (m):", model.coef_[0][0]) #we use [0][0] because coef_ is a 2D array
print("Intercept (b):", model.intercept_[0])

print("Mean Squared Error:", mean_squared_error(y, model.predict(x))) # we can see how much better scikit-learn performs than our scratch implementation

plt.scatter(x, y, color='blue')
plt.plot(x, model.predict(x), color='green')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary vs Years of Experience")
plt.show()






```
